pour calculer les données centrées réduites

```python
import pandas as pd

# -----------------------------
# 1. Données initiales
# -----------------------------
data = {
    "Ville": ["Casablanca", "Rabat", "Marrakech", "Fès", "Tanger",
              "Agadir", "Oujda", "Meknès", "Essaouira", "Nador"],
    "Disponibilité": [4.9, 4.7, 4.6, 4.8, 4.5, 4.9, 4.7, 4.6, 4.8, 4.5],
    "Courtoisie":    [4.7, 4.5, 4.4, 4.2, 4.3, 4.7, 4.9, 4.4, 4.6, 4.3],
    "Choix":         [2.4, 2.5, 2.6, 2.3, 2.7, 2.6, 2.9, 2.6, 2.5, 2.7],
    "Prix":          [2.6, 2.7, 2.6, 2.5, 2.9, 3.0, 3.1, 3.2, 2.7, 2.9]
}

df = pd.DataFrame(data)

# On met la colonne Ville comme index
df = df.set_index("Ville")

# -----------------------------
# 2. Moyennes
# -----------------------------
moyennes = df.mean()

# -----------------------------
# 3. Écarts-types
# ddof=1 pour l'écart-type d'échantillon
# -----------------------------
ecarts_types = df.std(ddof=1)

# -----------------------------
# 4. Données centrées réduites
# -----------------------------
df_cr = (df - moyennes) / ecarts_types

# Arrondi à 3 décimales
df_cr = df_cr.round(3)

# -----------------------------
# 5. Affichage
# -----------------------------
print("Moyennes :")
print(moyennes.round(3))
print("\nÉcarts-types :")
print(ecarts_types.round(3))
print("\nDonnées centrées réduites :")
print(df_cr)
```
